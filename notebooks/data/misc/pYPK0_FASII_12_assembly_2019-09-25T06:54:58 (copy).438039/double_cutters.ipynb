## Double cutters

Search for restriction enzymes in our freezer that cut transcription unit (TU) vectors twice producing at least two bands
from all vectors.

```python
from pydna.all import *
```

```python
seqnames = """\
pYPK0_ENO2_Athmod1_RPL5.gb
pYPK0_FBA1_EcfabG_RPL22A.gb
pYPK0_PDC1_EcfabH_TEF1.gb
pYPK0_PMP3_EcfabZ_ENO2.gb
pYPK0_RPL5_AthfatA1_RPL16A.gb
pYPK0_RPL16A_AthfatB_RPL17A.gb
pYPK0_RPL16B_EcacpS_TMA19.gb
pYPK0_RPL17A_EcacpH_RPL16B.gb
pYPK0_RPL22A_EcacpP_TDH3.gb
pYPK0_TDH3_EcfabF_UTR2.gb
pYPK0_TEF1_EcfabD_FBA1.gb
pYPK0_TPI1_EcfabA_PMP3.gb
pYPK0_UTR2_EcfabB_TPI1.gb""".splitlines()
```

```python
seqs = [read(s) for s in seqnames]
```

```python
from Bio.Restriction import (Eco31I,
 HaeI,
 ScrFI,
 AlwNI,
 BsaI,
 ScaI,
 Bsp119I,
 Bme1390I,
 BsgI,
 BstBI,
 BsuRI,
 Eco147I,
 PstI,
 HaeII,
 StuI,
 HaeIII,
 HindII,
 AseI,
 MnlI)
```

```python
rb = [Eco31I,
 HaeI,
 AlwNI,
 BsaI,
 ScaI,
 Bsp119I,
 BsgI,
 BstBI,
 Eco147I,
 PstI,
 HaeII,
 StuI,
 HindII,
 AseI,]
```

```python
[(e,e.size) for e in rb]
```

```python
for enz in rb:
    print(enz)
    for s in seqs:
        print([len(x) for x in sorted( s.cut(enz), key=len, reverse=True)])
```
